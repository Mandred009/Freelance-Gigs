```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_df=pd.read_csv('/content/drive/MyDrive/Titanic Kraggle/train.csv')
train_df.head()
```

```python
train_df.isnull().sum()
```

```python
# So we need to optimize data by filling nulls and removing unneccessary columns.
train_df=train_df.drop(columns=['Cabin','Ticket','Name','PassengerId'],axis=1)
train_df
```

```python
# To fill missing values in Age we will use mean filling
train_df['Age'].fillna(train_df['Age'].mean(), inplace=True)

# To fill missing values in the embarked column we will use mode of the values
print(train_df['Embarked'].mode())
train_df['Embarked'].fillna(train_df['Embarked'].mode()[0], inplace=True)
```

```python
train_df
```

```python
# Now after filling all nulls we will convert male and female column into categorical values ie 0 or 1 same for Embarked
train_df.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
```

```python
# Train Test Splitting Data
X = train_df.drop(columns= 'Survived',axis=1)
Y = train_df['Survived']

from sklearn.model_selection import train_test_split
from sklearn import preprocessing

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.55, random_state=2)
X_test.describe,X_train.describe
```

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model=LogisticRegression()
```

```python
# Fitting
model.fit(X_train,Y_train)
```

```python
# Prediction and accuracy
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

model_li=LinearRegression()
```

```python
# Fitting
model_li.fit(X_train,Y_train)
```

```python
# Prediction and accuracy
X_test_prediction_li = model_li.predict(X_test)
print(np.sqrt(np.mean(np.square(Y_test - X_test_prediction_li))))
```

```python
input=np.array((1,1,38.0,1,0,71.2500,1))
input=input.reshape(1,-1)
input
```

```python
# Prediction using custom data
print(model_li.predict(input))
print(model.predict(input))
```

# Support Vector Machine

```python
from sklearn import svm
clf = svm.SVC(kernel='rbf')
clf.fit(X_train, Y_train) 
```

```python
# Prediction
yhat = clf.predict(X_test)
yhat [0:5]
```

```python
# Evaluation using f1 score
from sklearn.metrics import f1_score
f1_score(Y_test, yhat, average='weighted') 
```

# Naive Byes Model

```python
from sklearn.naive_bayes import GaussianNB
model_nb=GaussianNB()
model_nb.fit(X_train,Y_train)
```

```python
# Prediction
pred_nb=model_nb.predict(X_test)
```

```python
# Evaluation
from sklearn.metrics import f1_score
f1_score(Y_test, pred_nb, average='weighted') 
```
